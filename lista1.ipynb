## Exercício 1.2
Carregue a imagem da Lena de dimensões 256x256 (com o adread) e apresente (com o adshow). Com esta imagem, faça um duplo loop para rotacionar a imagem de 90 graus para a esquerda. Meça o tempo que o sistema levará usando a função time. Por exemplo:

```python
import time

lena = imread('../images/images256/lena.png')
gray()
t1 = time.time()
lena_cw = zip(*lena[::-1])
# lena_ccw = zip(*lena)[::-1]
lena_ccw = zeros(shape(lena)) #[[0 for _ in range(len(lena))] for _ in range(len(lena[0]))]

h = len(lena)
w = len(lena[0])

for i in range(h):
    for j in range(w):
        lena_ccw[w - 1 - j][i] = lena[i][j]

t2 = time.time()
# pylab.rcParams['figure.figsize'] = (15.0, 8.0)

subplot(1,2,1)
imshow(lena)

subplot(1,2,2)
imshow(lena_ccw)

print "Tempo total =", t2-t1
```

## Exercício 1.4
Repita o exercício 1.2 usando a função meshgrid. Existe alguma vantagem de usar o meshgrid?

```python
import time
lena = imread('../images/images256/lena.png')
gray()
imshow(lena)

t1 = time.time()

h, w = shape(lena)
x, y = meshgrid(range(h),range(w))
lena_ccw_mg = lena[x, y * -1]

t2 = time.time()    
imshow(lena_ccw_mg)

print "Tempo total =", t2-t1
```

## Exercício 1.5
Com esta mesma imagem, calcule e apresente a média e o desvio padrão usando loops (range). Em seguida, calcule e apresente usando as funções do numpy. Existe alguma diferença?

Média:
$$
\mu =\frac { 1 }{ H } \sum _{ \upsilon = 0  }^{ { \upsilon  }_{ max } } \left( \upsilon { H }_{ \upsilon  } \right) 
$$

Desvio padrão:
$$
\sigma =\sqrt { \frac { \sum _{ \upsilon =0 }^{ { \upsilon  }_{ max } } \left[ { \left( \upsilon -\mu  \right)  }^{ 2 }H _{ \upsilon  } \right]  }{ N - 1 }  } 
$$

```python
lena = imread('../images/images256/lena.png')
gray()
imshow(lena)

a1 = time.time()

amostra = size(lena)

# media = sum(lena) / size(lena)
media = 0.0

for i in range(shape(lena)[0]):
    for j in range(shape(lena)[1]):
        media += lena[i][j]
media /= amostra
    
desvio_sum = 0.0
for i in range(shape(lena)[0]):
    for j in range(shape(lena)[1]):
        desvio_sum += (lena[i][j] - media)**2
        
desvio = sqrt(desvio_sum / amostra) # N-1 se estivermos num histograma local (que não seja a amostra inteira)

a2 = time.time()

# Com NumPy

b1 = time.time()

media2 = lena.mean()
desvio2 = lena.std()

b2 = time.time()

print "### Sem NumPy ###"
print "Media:         ", media
print "Desvio-padrão: ", desvio
print "Tempo total:   ", a2 - a1
print ""
print "### Com NumPy ###"
print "Media:         ", media2
print "Desvio-padrão: ", desvio2
print "Tempo total:   ", b2 - b1
```
