### Exercício 1
A imagem abaixo foi gerada com a função iacos da toolbox ia636. 
Leia a documentação da função e explique a imagem gerada pelo
comando `iacos(s,t,theta,0)`

```python
from ia636 import *

s = (256,256)
t = 64
theta = pi/2
f = iacos(s, t, theta, 0)
imshow(f)
gray()
```

#### Resposta 1
    
O método `iacos()` cria uma imagem em tons de cinza com uma representação cosenoidal com amplitude $1$ na dimensão `s`, com o período `t`. A orientação da imagem é ajustada com o parâmetro $\theta$ e offset com o parâmetro $\phi$

Internamente a imagem é gerada através da equação abaixo para cada $x$ e $y$:

$$
I_{ x,y }=\cos ({ 2\pi [(\frac { 1 }{ t } \cos { \theta  } )x+(\frac { 1 }{ t } \sin { \theta  } )y]+\phi  })
$$
