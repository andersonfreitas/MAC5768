## Exercício 2.1
Escreva uma função em Python que, dada uma imagem, retorne o histograma das frequências absolutas. Use essa função para calcular os histogramas de 5 imagens da lista de imagens da disciplina. Apresente (graficamente) os histogramas usando a função plot do módulo matplotlib. Compare o resultado da sua implementação com a da toolbox ia636.

```python
from skimage.io import imread
import ia636

pylab.rcParams['figure.figsize'] = (8.0, 15.0)

def histograma(img):
    histo = [0]*256
    
    for i in range(shape(img)[0]):
        for j in range(shape(img)[1]):
            histo[img[i][j]] += 1
    return histo

img1 = imread('../images/images256/lena.png')
hist1a = histograma(img1)
hist1b = ia636.iahistogram(img1)
subplot(5,2,1); plot(hist1a)
subplot(5,2,2); plot(hist1b)

img2 = imread('../images/images256/leopard.png')
hist2a = histograma(img2)
hist2b = ia636.iahistogram(img2)
subplot(5,2,3); plot(hist2a)
subplot(5,2,4); plot(hist2b)

img3 = imread('../images/images256/indor2.png')
hist3a = histograma(img3)
hist3b = ia636.iahistogram(img3)
subplot(5,2,5); plot(hist3a)
subplot(5,2,6); plot(hist3b)

img4 = imread('../images/images256/fiore.png')
hist4a = histograma(img4)
hist4b = ia636.iahistogram(img4)
subplot(5,2,7); plot(hist4a)
subplot(5,2,8); plot(hist4b)

img5 = imread('../images/images256/galaxia.png')
hist5a = histograma(img5)
hist5b = ia636.iahistogram(img5)
subplot(5,2,9); plot(hist5a)
subplot(5,2,10); plot(hist5b)
```

## Exercício 2.2
Escreva uma função em Python que, dada uma imagem, retorne o histograma das frequências relativas. Use essa função para calcular os histogramas das mesmas 5 imagens que usou no exercício anterior. Apresente (graficamente) os histogramas.

```python
def histograma_rel(img):
    histo = [0.0] * 256
    
    for i in range(shape(img)[0]):
        for j in range(shape(img)[1]):
            histo[img[i][j]] += 1
    
    N = size(img)

    for v in range(len(histo)):
        histo[v] = histo[v] / N

    return histo

img1 = imread('../images/images256/lena.png')
hist1a = histograma(img1)
hist1b = histograma_rel(img1)

pylab.rcParams['figure.figsize'] = (15.0, 5.0)
subplot(2,2,1)
plot(hist1a)
subplot(2,2,2)
plot(hist1b)
```
